First, we need to install required packages. Make a new environment with pip or conda. In terminal, type:

conda create -n 'DAS_IO'
conda activate DAS_IO

Or:

pip -m venv venv
venv/bin/Activate.ps1

Install package and its dependencies. If you are inside the git repository, use:

pip install -e .

Otherwise, use:

pip install git+https://github.com/snyderer/DAS_IO.git
